```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
```

```python
firstScatter = figure(height=400, width=400,
                      x_axis_label='X-values',
                      y_axis_label='Y-values')
```

```python
firstScatter.diamond(x=[1,2,3], y=[5,6,7], size=[10,20,15], color=['red','blue','pink'])

output_notebook()
show(firstScatter)
```

```python
fruits=['apples','bananas','oranges']
my_vbar = figure(x_range=fruits, height=400, width=400)
my_vbar.vbar(x=fruits, top=[20,10,50], width=0.5)
show(my_vbar)
```
