```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, NumeralTickFormatter
from bokeh.io import output_notebook, show
```

# In a new .py file, create a chart with time-series data showing total amount of contributions over time
Update this program to allow viewer to view contributions over time by selecting recipient name from a dropdown 


```python
contrib = pd.read_excel('data/2017_Contributions.xlsx')
```

```python
contrib.columns
```

```python
# Create a list with all the names of the recipients
recipList = contrib.RECIPNAME.unique()
# showing the first name in the list
recipList[0]
```

```python
# Slicing the dataframe to select only those records where recipient name matches. 
# In the code below recipList[0] would be replaced by the value user selects in the SELECT widget as mySelect.value
recipSelect = contrib[contrib['RECIPNAME'] == recipList[0]]
recipSelect
```

```python
# Now using groupby statement retrieve the funds raised over time on this sliced dataframe
recipAmnt = recipSelect.groupby(['DATE'])['AMNT'].sum()
recipAmntCDS = ColumnDataSource(data=dict(
                x_val = recipAmnt.index.values,
                y_val = recipAmnt.values
))
```

```python
myLine = figure(x_axis_type='datetime',
               width=600, height=400)

myLine.line(x='x_val', y='y_val', source=recipAmntCDS, line_color='blue')
myLine.yaxis.formatter = NumeralTickFormatter(format='0,0')

output_notebook()
show(myLine)
```
